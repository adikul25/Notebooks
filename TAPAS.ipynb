```python
import pandas as pd
from transformers import pipeline
```

```python
nlq_tapas = pipeline(task="table-question-answering",
                     model="google/tapas-base-finetuned-wtq")
```

```python
df = pd.read_csv('/content/port contracts.csv')
df = df.astype(str)
```

```python
df1 = pd.read_csv('/content/choclate_11.csv')
df1 = df1.astype(str)
```

```python
query = "what is the maximum fare  for 20 foot standard container and which POL country does it belong to?"
answer = nlq_tapas(table=df, query=query)['answer']
print(answer)
```

```python
questions = [
    "how many contract are open from India to Germany?",
    "Which countries are the Port of Loading (POL) and Port of Discharge (POD) for the shipment from BEIJIAO, China to Zeebrugge, Belgium?",
    "What is the LSS (Low Sulfur Surcharge) amount for the shipment from CHATTOGRAM,Bangladesh to Zeebrugge, Belgium?",
    "What is the validity date for the contract with contract number FRA2L01RC01?",
    "What is the average freight charge for a 40-foot high cube container from CHATTOGRAM, Bangladesh to Zeebrugge, Belgium?",
    "What is the total freight charge for a 20-foot Standard Container and a 40-foot high cube container from CHATTOGRAM, Bangladesh to Zeebrugge, Belgium?",
    "Is the Service Mode for the shipment from BEIJIAO, China to Zeebrugge, Belgium 'OCEAN TO OCEAN' or 'RAMP TO OCEAN'?",
    "Is the Service Mode for the shipment from LAT KRABANG, Thailand to Zeebrugge,Belgium 'OCEAN TO OCEAN' or 'RAMP TO OCEAN'?",
    "Which ports in China have a shipping cost to Belgium below $1200 for a 20-foot Standard Container?",
    "What is the validity date for the pricing information in the dataset?",
    "What is the currency used for shipping from YANTIAN, China to Le Havre, France?",
    "What is the average low-sulfur surcharge for shipping from all ports to Belgium?",
    "Which ports have a freight cost of $1300 from China to France for a 40-foot Standard Container?",
    "Which ports in Malaysia have a shipping cost below $1100 to Belgium for a 40-foot High Cube Container?",
    "Which ports in South Korea have a shipping cost below $1400 to Germany for a 20-foot High Cube Container?",
    "What is the shipping cost from Myanmar to Antwerp, Belgium for a 20-foot Standard Container?"
]
```

```python
for question in questions:
    answer = nlq_tapas(table=df, query=question)['answer']
    print("Question:", question)
    print("Answer:", answer)
    print()
```
